### Create KPIs

```python
gold_df = spark.sql("""
SELECT
symbol,
AVG(close_price) avg_price,
SUM(volume) total_volume,
AVG(open_price) avg_open_price,
AVG(high_price) avg_high_price,
AVG(low_price) avg_low_price
FROM silver_stock
GROUP BY symbol
""")
```

### Write

```python
gold_df.write \
.mode("overwrite") \
.format("delta") \
.option("overwriteSchema", "true") \
.save(
"s3://stock-market-data-charan/gold"
)
```

### SQL
### Create table

```python
%sql
CREATE TABLE IF NOT EXISTS gold_stock_metrics
USING DELTA
LOCATION 's3://stock-market-data-charan/gold';
```
