### Read


```python
bronze_df = spark.read.table(
    "bronze_stock"
)
```

### Transform

```python
from pyspark.sql.functions import *

silver_df = (
    bronze_df
        .dropDuplicates()
        .withColumn(
            "timestamp",
            to_timestamp("timestamp")
        )
)
```

### Write

```python
silver_df.write \
.mode("overwrite") \
.format("delta") \
.option("mergeSchema", "true") \
.save(
"s3://stock-market-data-charan/silver"
)
```

### SQL

```python
%sql
CREATE TABLE IF NOT EXISTS silver_stock
USING DELTA
LOCATION 's3://stock-market-data-charan/silver';
```
